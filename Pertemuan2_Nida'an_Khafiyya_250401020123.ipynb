# Struktur Data Python

## List

```python
data = [10, 20, 'Python', True, 3.14]

# Akses & slicing
print("List data", data)
print("Data index ke-0:", data[0]) # → 10
print("Data dengan Index paling belakang:", data[-1]) # → 3.14 (dari belakang)
print("List data ke-1 dan ke-3", data[1:3]) # → (20, 'Python')

# Manipulasi
data.append(99) # tambah di akhir
data.insert(1, 55) # sisipkan di posisi 1
data.remove(20) # hapus nilai pertama = 20
print("\nList yang telah dimanipulasi", data)
```

# Numpy: Komputasi Numerik

## NumPy Array

```python
# Dari list Python
a1D = np.array([1, 2, 3, 4, 5]) # 1D
a2D = np.array([[1,2,3],[4,5,6]]) # 2D

# Array dengan nilai inisial
zeros = np.zeros((3, 4)) # nol semua
ones = np.ones((2, 3)) # satu semua
eye = np.eye(4) # matriks identitas
full = np.full((2, 2), 7) # nilai konstan
print("Zeros\n", zeros, "\nOnes\n", ones, "\nEye\n", eye, "\nFull\n", full)

# Sequence
arange = np.arange(0, 20, 2) # 0,2,4,...18
linsp = np.linspace(0, 1, 11) # 11 titik merata
print("\nArange\n", arange, "\nLinsp\n", linsp)

# Array acak (reproducible)
rng = np.random.default_rng(seed=42)
rand_int = rng.integers(0, 100, (3, 4))
rand_flt = rng.normal(loc=0, scale=1, size=(100,))
print("\nRand Int\n", rand_int, "\nRand Flt\n", rand_flt)
```

## Operasi Vektorisasi dan Fungsi Statistik

```python
# Operasi aritmetika vektorisasi
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])
print(a + b) # → [11 22 33 44 55]
print(a * b) # → [10 40 90 160 250]
print(a ** 2) # → [1 4 9 16 25]

# Universal functions (ufuncs)
np.sqrt(a) # akar kuadrat tiap elemen
np.exp(a) # eksponensial (e^x)
np.log(b) # logaritma natural

# Fungsi statistik
data = rng.normal(0, 1, 1000)
print(f'\nMean: {np.mean(data):.4f}')
print(f'Std: {np.std(data):.4f}')
print(f'Min: {np.min(data):.4f}')
print(f'Max: {np.max(data):.4f}')
print(f'Q1: {np.percentile(data,25):.4f}')
print(f'Q3: {np.percentile(data,75):.4f}')
```

# Pandas: Manipulasi DataFrame

## Pandas Series

```python
# Series dari list — index default 0,1,2...
s1 = pd.Series([85, 92, 78, 95, 88])

# Series dengan index kustom
nilai = pd.Series(
[85, 92, 78, 95],
index=['Alice','Bob','Caca','Dina']
)
print("Nilai Bob:", nilai['Bob']) # → 92
print("Filter nilai:\n", nilai[nilai>80]) # filter
print("Nilai rata-rata:", nilai.mean()) # → 87.5
print("\nStatistik:\n", nilai.describe()) # statistik lengkap
```

# Praktikum: Dataset Titanic

```python
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(URL)
df.head()
```

## Q1: Baris dan Kolom

```python

df.info()
print("\nJumlah Baris dan Kolom:\n", df.shape)
print("Kolom:\n", df.columns.tolist())
```

Pada dataset ini, berisi **891 baris** dengan **total kolom yang berjumlah 12**. Adapun kolom yang tersedia, yaitu **PassengerId, Survived, Pclass, Name, Sex, Age, SibSp, Parch, Ticket, Fare, Cabin, dan Embarked**.

## Q2: Missing Value

```python
# missing values
missing_count = df.isnull().sum()
print(missing_count[missing_count > 0])

# persentase
missing_pct = (df.isnull().sum() / len(df)) * 100
print("\n", missing_pct[missing_pct > 0])
```

Pada kolom dataset tersebut, ada beberapa yang mengalami missing value. Kolom **Age** mengalami missing value sebanyak **177 baris (19.7%),** kolom **Cabin** sebanyak **687 baris (77.1%),** dan terakhir **Embarked** sebnyak **2 baris (0.2%).**

## Q3: Penumpang Survived

```python
persen_selamat = df['Survived'].mean() * 100
print(f"Persentase yang selamat: {persen_selamat:.2f}%")
```

Persentase penumpang yang berhasil selamat **(Survived)** dari dataset ini sebesar **38.38%**.

## Q4: Penumpang Wanita

```python
wk1 = df[(df["Sex"] == "female") & (df["Pclass"] == 1)]
print("Jumlah wanita dari kelas 1:", len(wk1),"\nRata-rata umur wanita dari kelas 1", wk1['Age'].mean())
```

Berdasarkan data, penumpang wanita kelas 1 berjumlah **94 orang** dengan rata-rata berumur **35 tahun**.

## Q5: Tingkat Keselamatan per Kelas

```python
hasil = df.groupby("Pclass").agg(
    total=('PassengerId', 'count'),
    selamat=('Survived', 'mean'),
).round(3)
print(hasil)
```

Berdasarkan hasil groupby, **kelas 1** menempati tingkat Survived yang paling banyak dengan rata-rata **63% (216 orang)**. Kemudian disusul oleh **kelas 2** dengan rata-rata **47% (184 orang)** dan terakhir **kelas 3** dengan rata-rata **24% (491 orang)**.

## Q6: Analisis Tambahan

```python
# Membuat kolom FamilySize dan status IsAlone
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = df['FamilySize'].apply(lambda x: 1 if x == 1 else 0)

# Analisis tingkat keselamatan
analisis_keluarga = df.groupby('IsAlone').agg(
    total=('PassengerId', 'count'),
    selamat=('Survived', 'mean')
).round(3)

print(analisis_keluarga)
```

Berdasarkan hasil analisis, penumpang yang bepergian bersama keluarga **(IsAlone=0)** memiliki tingkat keselamatan sebesar **50.6% (354 orang)**. Sedangkan penumpang yang sendirian **(IsAlone=1)** tingkat keselamatannya jauh lebih rendah, yaitu hanya **30.4% (537 orang)**.

Data ini menunjukkan kalau penumpang yang membawa keluarga punya peluang selamat lebih tinggi. Hal ini kemungkinan karena prioritas evakuasi "wanita dan anak-anak" yang biasanya bepergian dalam grup keluarga, atau adanya aksi saling bantu antar anggota keluarga saat mencari sekoci.
