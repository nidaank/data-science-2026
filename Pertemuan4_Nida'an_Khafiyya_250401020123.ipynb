## Load & Inspect Dataset

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns
from scipy import stats
```

```python
df = sns.load_dataset('iris')
print('Shape:', df.shape)
print(df.dtypes)
print(df.head())
print(df.describe().round(3))
```

Pada eksplorasi awal, dilakukan pengecekan dataset Iris dengan menampilkan ukuran data (`shape`), tipe data setiap kolom (`dtypes`), lima data teratas (`head()`), serta ringkasan statistik numerik (`describe()`). Dataset Iris memiliki 150 data dan 5 kolom yang terdiri dari 4 fitur numerik bertipe `float64`, serta 1 kolom kategorikal yaitu `species`. Hasil statistik deskriptif menunjukkan informasi seperti nilai minimum, maksimum, rata-rata, median, dan persebaran data yang berguna untuk memahami karakteristik dataset sebelum dilakukan analisis lebih lanjut.

Dataset ini terdiri dari tiga spesies bunga iris, yaitu `setosa`, `versicolor`, dan `virginica`. Menariknya, `setosa` dikenal paling mudah dibedakan karena memiliki ukuran petal yang jauh lebih kecil dibanding dua spesies lainnya, sedangkan `virginica` umumnya memiliki ukuran petal paling besar dan `versicolor` berada di antara keduanya.


## Statistik Deskriptif Lengkap

```python
for col_name in df.select_dtypes(include='number').columns:
  col = df[col_name]
  print(f'\n=== {col_name} ===')
  print(f' Mean : {col.mean():.3f}')
  print(f' Median : {col.median():.3f}')
  print(f' Std Dev : {col.std():.3f}')
  print(f' Skewness : {col.skew():.3f}')
  print(f' Kurtosis : {col.kurt():.3f}')
```

Code ini gunakan untuk menghitung statistik deskriptif lanjutan pada setiap kolom numerik dalam dataset Iris, yaitu nilai **mean**, **median**, **standar deviasi**, **skewness**, dan **kurtosis**. Hasil analisis menunjukkan bahwa seluruh fitur memiliki nilai skewness yang **mendekati 0** sehingga distribusi data cenderung **simetris**, sedangkan nilai kurtosis yang sebagian besar bernilai negatif menunjukkan **distribusi data lebih landai** dibanding distribusi normal. Selain itu, fitur petal_length memiliki standar deviasi paling besar yang menandakan **variasi data paling tinggi** dibanding fitur lainnya.

## Analisis Distribusi (Histogram + KDE)

```python
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(df['sepal_length'], kde=True,
            color='#028090', bins=20, ax=ax)
ax.axvline(df['sepal_length'].mean(), color='red',
          linestyle='--', label=f"Mean={df['sepal_length'].mean():.2f}")
ax.axvline(df['sepal_length'].median(), color='orange',
          linestyle='--', label=f"Median={df['sepal_length'].median():.2f}")
ax.set_title('Distribusi Sepal Length'); ax.legend(); plt.show()
```

Distribusi `sepal_length` cenderung mendekati **distribusi normal** karena bentuk histogram terlihat cukup simetris dan membentuk pola menyerupai lonceng (bell-shaped). Hal ini juga didukung oleh **nilai mean (5.84) yang sangat dekat dengan median (5.80)**, serta nilai skewness sebelumnya sebesar **0.315** yang mana menunjukkan interpretasi **Simetris/Normal** (dalam rentang -0.5 hingga 0.5).


## Boxplot per Spesies

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(data=df, x='species', y='petal_length',
           palette='Set2', ax=axes[0])
axes[0].set_title('Boxplot Petal Length')
sns.violinplot(data=df, x='species', y='petal_length',
              palette='Set2', inner='box', ax=axes[1])
axes[1].set_title('Violin Plot Petal Length')
plt.tight_layout(); plt.show()
```

Berdasarkan boxplot dan violin plot, spesies `virginica` memiliki nilai `petal_length` terbesar dengan **median dan rentang data paling tinggi** dibanding spesies lainnya. Sebaliknya, spesies `setosa` memiliki nilai `petal_length` terkecil dengan distribusi data yang terkonsentrasi pada **rentang rendah sekitar 1–2**. Sementara itu, spesies `virginica` juga terlihat sebagai spesies yang **paling terseba**r karena memiliki rentang distribusi dan variasi data paling luas dibanding `setosa` dan `versicolor`.

Selain itu, pada boxplot terlihat bahwa `setosa` memiliki **outlier** di bagian **atas maupun bawah distribusi**, sedangkan `versicolor` memiliki outlier pada bagian **bawah distribusi** data.


## Matriks Korelasi Pearson

```python
corr = df.drop('species', axis=1).corr(method='pearson')
print(corr.round(3))
import numpy as np
mask = np.triu(np.ones(corr.shape, dtype=bool))
corr_masked = corr.where(~mask)
max_pair = corr_masked.stack().idxmax()
print(f'Korelasi tertinggi: {max_pair} = {corr.loc[max_pair]:.3f}')
```

Pada hasil korelasi Pearson, pasangan kolom dengan korelasi tertinggi adalah `petal_width` dan `petal_length` dengan nilai korelasi sebesar **0.963**, yang menunjukkan hubungan positif sangat kuat dimana **semakin panjang petal maka lebar petal juga cenderung meningkat**. Sementara itu, pasangan kolom dengan korelasi terendah adalah `sepal_width` dan `petal_length` dengan nilai korelasi sebesar **-0.428,** yang menunjukkan hubungan negatif sedang dimana **peningkatan panjang petal cenderung diikuti penurunan lebar sepal**.

## Scatter Plot & Heatmap

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.scatterplot(data=df, x='petal_length', y='petal_width',
               hue='species', palette='Set2', ax=axes[0])
sns.regplot(data=df, x='petal_length', y='petal_width',
           scatter=False, color='gray', ax=axes[0])
axes[0].set_title('Scatter: Petal Length vs Petal Width')
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdBu_r',
           vmin=-1, vmax=1, square=True, ax=axes[1])
axes[1].set_title('Heatmap Korelasi')
plt.tight_layout(); plt.show()
```

Berdasarkan **scatter plot** `petal_length` terhadap `petal_width`, terlihat adanya hubungan **linear positif** yang kuat, dimana semakin besar nilai `petal_length` maka nilai `petal_width` juga cenderung meningkat. Persebaran data juga menunjukkan pemisahan spesies yang cukup jelas, dengan `setosa` berada pada nilai paling rendah, `versicolor` di tengah, dan `virginica` memiliki nilai paling tinggi. Garis regresi yang terus meningkat memperkuat adanya **korelasi positif** antar kedua variabel tersebut.

Pada **heatmap korelasi**, pasangan `petal_length` dan `petal_width` memiliki nilai korelasi tertinggi sebesar **0.96** yang menandakan hubungan sangat kuat dan searah. Sebaliknya, `sepal_width` memiliki korelasi negatif terhadap fitur petal, terutama dengan `petal_length` sebesar **-0.43**, yang menunjukkan bahwa peningkatan panjang petal cenderung diikuti **penurunan lebar sepal**. Secara keseluruhan, **fitur petal terlihat lebih dominan** dalam membedakan spesies bunga iris dibanding fitur sepal.
